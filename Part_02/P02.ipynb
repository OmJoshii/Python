## Global Variables

```python
x = " fun"

def myfunc():
    print("Python is" + x)

myfunc()
```

```python
#Create a variable inside a function, with the same name as the global variable
x = " awesome"

def mufunc():
    x = " fun"
    print("Python is" + x)

mufunc()

print("Python is" + x)
```

## The global Keyboard

```python
#To create a global variable inside a function, you can use the global keyword.

def myfunc():
    global x
    x = "fun"

myfunc()

print('Python is' + x)
```

```python
#To change the value of a global variable inside a function, refer to the variable by using the global keyword:

x = "fun"

def myfunc():
    global x
    x = " awesome"

myfunc()

print("Python is" + x)
```
